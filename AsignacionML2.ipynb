# 1. Data Science, Machine Learning, and Statistics: what is in a name?

### What is in a name?

Existen muchos artículos que se pueden encontrar en internet, los cuales están etiquetados como "Data Science" sin embargo es interesante y importante saber que hay detrás de este para de palabras, se enfatiza que cuando se menciona Data Science, se está refiriendo a la recopilación y gestión de datos, la aplicación de técnicas de modelado y la aplicación de estadísticas críticas, todo esto permitiendo el despliegue de modelos de datos valiosos en la producción.
Un ejemplo de esto es cuando un científico, que forma parte de un grupo de investigación, está trabajando en la simulación de la interacción entre ciertas moléculas biológicas, donde se utiliza una gran cantidad de datos, los cuales son muy valiosos en este proceso. Entonces cuando un trabajo y sus resultados están profundamente relacionado o depende fuertemente de los datos, entonces podríamos estar hablando de "Data Science".
Además se presenta un análisis de distintas ciencias de la información tales como estadística, Machine Learning, minería de datos, informática, Big Data, analítica predictiva y Data Science.

La estadística es el campo que trata con los datos con mayor portabilidad, lo que se refiere que no depende de un modelo físico. Las estadísticas generalmente se usan para saber que puede salir mal en un experimento y puede deducirse que errores pueden salir a partir del uso de un conjunto de datos, ya que poseen suficiente cantidad de técnicas que pueden resolver casi cualquier problema. Se puede decir que la mejor fuente de un buen trabajo estadístico son los malos experimentos ya que de ahí se obtienen ciertos errores que nos ayudarán a resolver problemas posteriores, si todos los experimentos se realizarían correctamente, no sería muy necesario el uso de estadísticas.

Otra característica importante es que es el único campo que enfatiza los riesgos de los datos pequeños, el poder razonar a partir de pequeños conjuntos de datos, ya que este tipo conjunto de datos son costosos de producir. En cambio con los datos grandes es un poco distinto el análisis ya que aveces se consideran valiosos porque son las más baratos de producir, su valor recae en el hecho que son un buen filtro para un conjunto de datos más pequeño, que como mencionamos anteriormente es costoso de producir, asi que si desea producir datos realmente valiosos directamente tendrá que trabajar con datos pequeños y consultar con un buen estadístico.

Machine learning es el campo que toma las riendas en aquel lugar donde los estadísticos temen entrar, esta rama tiene un concepto de lo que es trabajar con pequeños conjuntos de datos, sin embargo su principal objetivo es el crear modelos predictivos que sea indistinguible de un modelo correcto, lo cual es distinto respecto al pensamiento estadístico, ya que en este caso la estadística se preocupa por la creación de un modelo que sea exacto y además que sea correcto. Uno de los principales conceptos que se debe tener en cuenta cuando se habla de machine learning es que este está fuertemente relacionado con la optimización, entonces si se desea hacer un buen trabajo de machine learning se debe rehacer una predicción como un problema de optimización utilizando las técnicas oportunas para cada caso.

El término de data mining es aquel que ahora la mayoría de personas relacionan con data science, sin embargo muchas de esas personas confunden estos dos términos ya que en realidad tienen dos enfoques distintos. EL objetivo de Data mining es principalmente el encontrar relaciones entre un conjunto de datos, no necesariamente realizar prediciones, pueden haber muchas variables independientes y encontrar relaciones entre estas variables  mediante agrupamiento o detección de carácteristicas pero no hay variables dependientes.

La informática y bioinformática son términos que son importantes y de hecho sirven de mucho al momento de definir Data Science, si tomamos el término bioinformática y lo separamos en estadística y biología, nos quedamos con lo que es una infraestructura de datos y algoritmos de matching, además tenemos bases de datos y la creación y gestión de estos. 

Cuando se habla de Big Data, el cual es un término muy comentado en la actualidad en diversos campos de estudio, se refiere más que nada a la infraestructura o la palataforma superior en la cual se realiza el modelado, podemos mencionar MapReduce, Hadoop, noSQL, etc.

El área de analítica predictiva puede definirse como el conjunto de técnicas que provienen de otros campos tales como estadística, machine learning, minería de datos, etc; que son usadas sobre un conjunto de datos históricos actuales para realizar predicciones para el futuro. Es un término que suele estar relacionado con Data Science, pero que no es exactamente este, ya que en la analítica se trata más de visualización, informes y un resumen de un conjunto de datos determinado.

Ahora sí se menciona el tan esperado término, "Data Science", este campo usa varias técnicas y teoría de otros campos, tales como la matemática, ingeniería de datos, estadística, reconocimiento de patrones, etc; para la gestión de todo el proceso cuando se desea contruir un modelo, por esto se refiere al análisis empresarial, recopilación de datos, la administración de estos, la construcción y despliegue de modelos en producción.

# 3. A Plan for Spam

###¿Cuáles son los componentes clave de su sistema de filtrado estadístico? En otras palabras, ¿cómo funciona?.
Segun lo leido en el articulo el sistema de filtrado estadistico esta basado en la combinacion bayesianas de probabilidades de palabras individuales, inicialmente se escanea todo el corpus incluyendo encabezados, html y javascript incrustados en este, se escanea un correo spam y un no spam con la finalidad generar tokens. Se considera como parte de los tokens a los caracteres alfanumericos, los guiones, los apostrofes y signos de dolar pero aquellas palabras que son solo dígitos y los no nombrados separadores de tokens, del ultimo conjunto se ignora los comentarios html.

Una vez generados los tokens se empieza a contabilizar cada uno de estos en cada corpus para obtener su frecuencia para obtener dos tabla hash de ambos corpus y luego obtener una tercera tabla hash en la cual a cada token se le asigna la probabilidad de que un correo electronico que contiene este sea un spam, el calculo se realiza de la siguiente forma:
```
(let ((g (* 2 (or (gethash word good) 0)))
      (b (or (gethash word bad) 0)))
  (unless (< (+ g b) 5)
     (max .01
          (min .99 (float (/ (min 1 (/ b nbad))
                             (+ (min 1 (/ g ngood))   
                                (min 1 (/ b nbad)))))))))
```
Donde word es el token cuya probabilidad calculamos, good y bad son las tablas hash, ngood y nbad son la cantidad de mensajes no spam y spam respectivamente.
Como se observa en la linea 1 se duplica los números de la tabla hash god para sesgar ligeramente la probabilidad cuya finalidad es evitar falsos positivos, como se observa en la linea 3 solo se escogen las palabras que ocurren mas de cinco veces en total, esto habiendo ya duplicado los numeros de la tabla god; luego se le asigna probabilidad de 0.01  y 0.99 a las palabras que aparecen en un corpus pero no en el otro.

Cuando llega un nuevo correo, se sigue los pasos anteriores para tokenizar y se escoge los quince tokens más interesantes,de don se mide que su probabilidad de spam provenga de un .5 neutral y luego se utilizan para calcular la probabilidad de que el correo sea spam. Si probs es una lista de las quince probabilidades individuales, calculas la probabilidad combinada así:
```
(let ((prod (apply # '* probs)))
  (/ prod (+ prod (apply # '* (mapcar #' (lambda (x)
                                         (- 1 x))
                                     probs)))))
```

Según la extensión del Conjunto de Datos
------------------------------------------
El siguiente cuadro resume cuándo escoger un modelo clasificatorio según la extensión del dataset.
Cuando la data es pequeña se prefiere aumentar el bias y reducir la varianza, mientras para data grande se prefiere lo contrario, nacen los conceptos de clasificadores Generativos y Discriminativos.

| Extensión    | Bias (Sesgo) | Varianza | Clasificadores |
|:-------------|:------------:|:--------:|:---------------|
|Small dataset | alto         | baja     | Generativos    |
|Large dataset | bajo         | alto     | Discriminativos|

__1. Modelos Generativos:__

Los modelos generativos modelan la distribución de clases individuales, es decir cómo se generaron los datos para categorizar una señal. Hace la pregunta: "en función de los supuestos de mi generación, ¿qué categoría es más probable que genere esta señal?.

Modelos Generativos más populares:
* Gaussian mixture model and other types of mixture model
* Hidden Markov model
* Probabilistic context-free grammar
* Naive Bayes
* Averaged one-dependence estimators
* Latent Dirichlet allocation
* Restricted Boltzmann machine
* Generative adversarial networks

__2. Modelos Discriminativos:__

Los modelos discriminatorios aprenden el límite entre las clases, no le importa cómo se generaron los datos, simplemente categoriza una señal dada.

Modelos Discriminatívos más populares:
* Logistic regression
* Support Vector Machines
* Maximum Entropy Markov Model
* Conditional Random Fields
* Neural Networks
* Random forests
* k Nearest Neighbours

#### CASO 2
Considerando $(x,y) = \{ (0,0); (0,0); (1,0); (1,1) \}$, suponiendo que $x$ son los datos de entrada y $y$ los de salida, tenemos:
- Distribución de probabilidad conjunta $p(x,y)$.

|           | __y = 0__ | __y = 1__ |
|-----------|----------:|:---------:|
| __x = 0__ |    0.50   |  0.00     |
| __x = 1__ |    0.25   |  0.25     |
- Distribución de probabilidad condicional $p(y\mid x)$.

|           | __y = 0__ | __y = 1__ |
|-----------|----------:|:---------:|
| __x = 0__ |    1.00   |  0.00     |
| __x = 1__ |    0.50   |  0.50     |

__Análisis__: Notamos que $p(y\mid x)$ trata de aprender de la data directamente y luego clasificarla (_Discriminative_); mientras que $p(x,y)$ intenta aprender quien puede transformarse en $p(y\mid x)$ después de clasificar la data (_Generative_).

Ventajas de Algunos Modelos
------------------------------------------
__Naive Bayes:__

Un clasificador Naive Bayes convergerá más rápido que los modelos discriminativos si se cumple la suposición de independencia condicional.<br>
Necesita menos datos de entrenamiento y si la suposición de NB no se cumple, NBClassifier aún hace un gran trabajo en la práctica.<br>
Una buena apuesta si quieres algo rápido y fácil que funciona bastante bien. 

Desventaja: _No puede aprender interacciones entre características_.<br> 
Ejemplo: Bruno aunque ama las películas con Brad Pitt y Tom Cruise, odia las películas donde están juntos).

__Logistic Regression:__

Hay muchas formas de regularizar este modelo.<br>
No es necesario que sus características estén correlacionadas, como lo hace en Naive Bayes.<br>
Tiene una buena interpretación probabilística, a diferencia de los árboles de decisión o SVM.<br>
Modelo fácil de actualizar por gradiente de descenso en linea, a diferencia de los árboles de decisión o SVM.

Uso Recomendado:
- Cuando se busca marco probabilístico.
- Cuando se ajusta fácilmente los umbrales de clasificación.
- Para obtener intervalos de confianza seguridad de eventos.
- Cuando se espera incorporar nueva data al modelo.

__Decision Trees:__

Fácil manejo de interacciones de las características y no son paramétricas.<br>
Los datos son separables linealmente y valores atípicos casi no afectan la precisión.<br>
Plantea el problema para que todas las opciones sean analizadas.<br>
Permite analizar totalmente las posibles consecuencias de tomar una decisión.<br>
Permite cuantificar el costo de un resultado y la probabilidad de que suceda.

Desventaja:<br>
_No son compatibles con el aprendizaje en línea._<br>
_Requiere reconstrucción del modelo cuando se genera nueva data._<br>
_Suele tender al overfitting, solucionable con Random Forest._
Sólo es recomendable para cuando el número de acciones es pequeño y no son posibles todas las combinaciones

Random Forest suele ser una variante mejorada, es rápido y escalable.<br>
Además no requiere ajuste de parámetros como lo hace con SVM.

__SVMs:__

Lorga alta precisión y buenas garantías teóricas con respecto al overfitting.<br>
La selección de un kernel adecuado pueden optimizar el resultado incluso si la data no es linealmente separable en el espacio de la base.<br>
De buen rendimiento en problemas de clasificación de texto, donde los espacios altamente dimensionables son la norma.

Desventaja:<br>
Dificil de interpretar.<br>
Molesto de correr y optimizar.<br>

````python
print(__doc__)


# Code source: Gaël Varoquaux
#              Andreas Müller
# Modified for documentation by Jaques Grobler
# License: BSD 3 clause

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression

h = .02  # step size in the mesh

names = ["Naive Bayes", "Logistic Regression", "Decision Tree", "SVM"]

classifiers = [GaussianNB(), LogisticRegression(), DecisionTreeClassifier(), SVC()]

X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                           random_state=1, n_clusters_per_class=1)
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)
linearly_separable = (X, y)

datasets = [make_moons(noise=0.3, random_state=0),
            make_circles(noise=0.2, factor=0.5, random_state=1),
            linearly_separable
            ]

figure = plt.figure(figsize=(27, 9))
i = 1
# iterate over datasets
for ds_cnt, ds in enumerate(datasets):
    # preprocess dataset, split into training and test part
    X, y = ds
    X = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=.4, random_state=42)

    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    # just plot the dataset first
    cm = plt.cm.RdBu
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])
    ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
    if ds_cnt == 0:
        ax.set_title("Input data")
    # Plot the training points
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright,
               edgecolors='k')
    # and testing points
    ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6,
               edgecolors='k')
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xticks(())
    ax.set_yticks(())
    i += 1

    # iterate over classifiers
    for name, clf in zip(names, classifiers):
        ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)

        # Plot the decision boundary. For that, we will assign a color to each
        # point in the mesh [x_min, x_max]x[y_min, y_max].
        if hasattr(clf, "decision_function"):
            Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
        else:
            Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

        # Put the result into a color plot
        Z = Z.reshape(xx.shape)
        ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)

        # Plot also the training points
        ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright,
                   edgecolors='k')
        # and testing points
        ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright,
                   edgecolors='k', alpha=0.6)

        ax.set_xlim(xx.min(), xx.max())
        ax.set_ylim(yy.min(), yy.max())
        ax.set_xticks(())
        ax.set_yticks(())
        if ds_cnt == 0:
            ax.set_title(name)
        ax.text(xx.max() - .3, yy.min() + .3, ('%.2f' % score).lstrip('0'),
                size=15, horizontalalignment='right')
        i += 1

plt.tight_layout()
plt.show()
````

__Ten en cuenta que suponemos que los datos de los conjuntos de validación y prueba no se ven completamente cuando entrenamos nuestro modelo y, por lo tanto, no utilizamos ninguna información contenida en ellos.__

__Ignora todas las palabras que aparecen en menos de X = 30 correos electrónicos del conjunto de 4000 mensajes de correo electrónico, esto es tanto un medio de prevenir el ajuste excesivo y de mejorar la escalabilidad. Para cada correo electrónico, transformarlo en un vector de características $\mathbf{x}$ donde la i-ésima entrada, $x_i$, es 1 si la i-ésima palabra del vocabulario aparece en el correo electrónico y 0 en caso contrario.__

````python
import re      # expresiones regulares
import string

def frecuenciaPalabras(x_train):
    frequencyAll = {}
    correos = {}
    for i in range(len(x_train)):
        frequency = {} # frecuencia en correo i-esimo
        text_string = str(x_train.values[i])
        match_pattern = re.findall(r'\b[a-z]{4,12}\b', text_string) # expresion regular entre 4 y 12 caraacteres
        # frecuencia de aparicion de las expresion regular
        for word in match_pattern:
            count = frequency.get(word,0)
            frequency[word] = count + 1
        # llenado a diccionarios de frecuencias totales y numero de apariciones en correo
        for x in frequency.keys():
            if(x in correos.keys()): # contador de correos en los que aparece
                correos[x] = correos[x] + 1
            else:
                correos[x] = 1
            if(x in frequencyAll.keys()): # relleno de frecuencias totales de expReg
                frequencyAll[x] = frequencyAll[x] + frequency[x]
            else:
                frequencyAll[x] = frequency[x]
    # agregar [expReg numCorreos numApariciones]
    solution = []
    for words in frequencyAll.keys():
        if(correos[words] >=30):
            solution.append([words, correos[words], frequencyAll[words]])
    # parsear a DataFrame
    dataCompreso = pd.DataFrame(np.array(solution), columns=["Palabra","nCorreos","nRepeticiones"])
    dataCompreso["nCorreos"] = dataCompreso["nCorreos"].astype(int)
    dataCompreso["nRepeticiones"] = dataCompreso["nRepeticiones"].astype(int)
    return dataCompreso
    


"""
Fuente: 
https://code.tutsplus.com/tutorials/counting-word-frequency-in-a-file-using-python--cms-25965
https://pandas.pydata.org/pandas-docs/stable/merging.html
"""
frecuenciaPalabras(x_train)
````

__3 .Implementar las funciones **entrenamiento_perceptron(datos)** y **prueba_perceptron(w, datos)**.__

__La función `entrenamiento_perceptron(datos)` entrena un clasificador perceptron usando los ejemplos proporcionados a la función, y devuelve `w`, `k`, `iter`, el vector de clasificación final, el número de actualizaciones (errores) realizados y el número de pasadas los datos, respectivamente.__

__Se puede  suponer que los datos de entrada proporcionados a la función son linealmente separables (por lo que el criterio de parada debe ser que todos los puntos estén clasificados correctamente).__

__Para el caso  `w·x = 0`, predecir la clase +1 (spam). Para este ejercicio, no es necesario agregar una función de sesgo al vector de características__

__La función `prueba_perceptron(w, datos)` debe tomar como entrada el vector de peso `w`  (el vector de clasificación a utilizar) y un conjunto de ejemplos. La función debe devolver el error de prueba, es decir, la fracción de ejemplos que son mal clasificados por w.__

````python
import numpy as np
import random
### Clase Perceptrón ###
class Perceptron:
    """
    Constructor
    """
    def __init__(self, vectores=None, grupos=None):
        ### Variables ###
        self.vectores=[]
        self.grupos=[]
        self.vClasi=[] ### Vector de pesos o de clasificación
        self.maxIteraciones=0
        
        self.setVectores(vectores, grupos)
        """
        Establecer los vectores y los grupos(spam (1)o no spam(0))
        """
    def setVectores(self, vectores=None, grupos=None):
        if vectores is not None:
            self.vectores=vectores
            self.vClasi=np.zeros(len(self.vectores[0]),int)
            #self.vClasi = [random.random() for _ in range(len(self.vectores[0]))]
        if grupos is not None:
            self.grupos=grupos

    def clasificador(self, vector):
        y=np.dot(vector, self.vClasi)
        if y>=0:
            return 1
        return -1

    def prueba_perceptron(self, vectores, grupos): ### Comun para todas las clases
        erroneos=0
        for i in range(len(vectores)):
            y=self.clasificador(vectores[i])
            if(y==1 and grupos[i]==0) or (y==-1 and grupos[i]==1):
                erroneos=erroneos+1

        return float(erroneos)/float(len(vectores))

class Perceptron_normal(Perceptron):
    def entrenamiento_perceptron(self, vectores=None, grupos=None):
        self.setVectores(vectores, grupos)
        errores=0
        iteraciones=0

        #Ejecuta las iteraciones
        while self.maxIteraciones == 0 or self.maxIteraciones-iteraciones>0:
            
            iteraciones=iteraciones+1
            temp=errores

            #Recorrer todos los vectores
            for vect in range(len(self.vectores)):
                y=self.clasificador(self.vectores[vect])
                
                #Clasificaciones incorrectas
                if(y==1 and self.grupos[vect]==0) or (y==-1 and self.grupos[vect]==1):
                    self.vClasi=np.add(self.vClasi, np.multiply(-y, self.vectores[vect]))
                    errores=errores+1
                

            if temp == errores:
                break

        return self.vClasi, errores, iteraciones
````

__6 . Implementar el algoritmo de perceptron promedio, que es el mismo que tu implementación actual, pero que, en lugar de devolver el vector de peso final, devuelve el promedio de todos los vectores de peso considerados durante el algoritmo (incluyendo ejemplos donde no se cometió ningún error). El promedio reduce la varianza entre los diferentes vectores, y es un poderoso medio para evitar que el algoritmo de aprendizaje sea sobrefijado (sirviendo como un tipo de regularización).__

````python
import numpy as np
### Clase Perceptrón ###
class Perceptron:
    """
    Constructor
    """
    def __init__(self, vectores=None, grupos=None):
        ### Variables ###
        self.vectores=[]
        self.grupos=[]
        self.vClasi=[]  ### Vector de pesos o de clasificación
        self.maxIteraciones=0
        self.setVectores(vectores, grupos)
        """
        Establecer los vectores y los grupos(spam (1)o no spam(0))
        """
    def setVectores(self, vectores=None, grupos=None):
        if vectores is not None:
            self.vectores=vectores
            self.vClasi=np.zeros(len(self.vectores[0]),int)
        if grupos is not None:
            self.grupos=grupos

    def clasificador(self, vector):
        y=np.dot(vector, self.vClasi)
        if y>=0:
            return 1
        return -1

    def prueba_perceptron(self, vectores, grupos): ### Comun para todas las clases
        erroneos=0
        for i in range(len(vectores)):
            y=self.clasificador(vectores[i])
            if(y==1 and grupos[i]==0) or (y==-1 and grupos[i]==1):
                erroneos=erroneos+1

        return float(erroneos)/float(len(vectores))

### Clase del Perceptrón promedio ###
class Perceptron_prom(Perceptron):
    def entrenamiento_perceptron(self, vectores=None, grupos=None):
        self.setVectores(vectores, grupos)
        errores=0
        iteraciones=0

        #Vector para procesar el promedio
        totalvector=list(self.vClasi)
        numUsados=1

        #Ejecutar iteraciones
        while self.maxIteraciones == 0 or self.maxIteraciones-iteraciones>0:
            iteraciones=iteraciones+1
            temp=errores

            #Recorrer todos los vectores
            for vect in range(len(self.vectores)):
                y=self.clasificador(self.vectores[vect])

                #Clasificaciones incorrectas
                if(y==1 and self.grupos[vect]==0) or (y==-1 and self.grupos[vect]==1):
                    self.vClasi=np.add(self.vClasi, np.multiply(-y, self.vectores[vect]))
                    errores=errores+1

                #Agregar al promedio
                totalvector=np.add(self.vClasi, totalvector)
                numUsados=numUsados+1

            if temp==errores:
                self.vClasi=np.multiply(1.0/(numUsados), totalvector)
                break

        return self.vClasi, errores, iteraciones
    
````
